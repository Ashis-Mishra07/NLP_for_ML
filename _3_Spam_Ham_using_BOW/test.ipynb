## BOW , TFIDF , Marchine Learning Algorithms

```python
import pandas as pd
messages = pd.read_csv('file.csv', sep='\t', names=["label", "message"])
```

```python
messages
```

```python
import re    # regular expression
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []

for i in range(0 ,len(messages)):
    review = re.sub('[^a-zA-Z]' , ' ' , messages['message'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

#### BOW - Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500 , ngram_range=(1,2))
```

```python
# Independent Features
X = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30 ,linewidth=100000 , formatter=dict(float=lambda x: "%.3f" % x))
X
```

```python
cv.vocabulary_
```

```python
# Output or Dependent Features
y = pd.get_dummies(messages['label'])
```

```python
y = y.iloc[:,0].values
```

```python
y
```

```python
# Train and test split
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.20 , random_state=30)
```

```python
# X_train and y_train are my independent features and dependent features
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train , y_train)
```

```python
spam_detect_model 
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
# to compare
from sklearn.metrics import accuracy_score , classification_report
accuracy = accuracy_score(y_test , y_pred)   # this will get compare but we will analyse it with classification reports 
```

```python
accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred))
```

- But the mistake here in above program is , we have done train_test_split after BOW and TFIDF and it is to be done in before steps .

- This is beacuse if we do train_test_split in first , then model have no idea regarding the test and the correct prediction will be taken .
- Now if we do TF-IDF , some data regrading the test is already known so the prediction percentage will be high but it will fail in rel world problems .
- By doing this we are preventing the data leakage .
